```python
#1
num=int(input("Enter the number:"))
def add(n):
    if(n<=0):
        print("the given number is not a natural number")
    elif(n==1):
        return 1
    else:
        return(n+add(n-1))
print(f"the sum of a {num} natural numbers",add(num))
```

```python
#2
words=input("enter the string:").split()
def create_Dict(words):
    Dict={}
    for word in words:
        if word:
            first_char=word[0].upper()
        if first_char not in Dict:
            Dict[first_char]=[]
        Dict[first_char].append(word)
    return Dict
result=create_Dict(words)
sort=sorted(result.items())
for key,value in sort:
    print(f"{key}:{value}")
```

```python
#3
def count_char(string):
    char_count={}
    for char in string:
        if char.isalpha():
            if char in char_count:
                char_count[char]+=1
            else:
                char_count[char]=1
    return char_count
string=input("enter the string:")
Dict=count_char(string)
print("characters count in string:")

print(Dict)
```

```python
#6
class ComplexNumber:
    def __init__(self,real,imaginary):
        self.real=real
        self.imaginary=imaginary
        
    def __add__(self,other):
        return ComplexNumber(self.real+other.real,self.imaginary+other.imaginary)
    
    def __sub__(self,other):
        return ComplexNumber(self.real-other.real,self.imaginary-other.imaginary)
  
    def __mul__(self,other):
        real_part=(self.real*other.real)-(self.imaginary*other.imaginary)
        imaginary_part=(self.real*other.imaginary)-(self.imaginary*other.real)
        
        return ComplexNumber(real_part,imaginary_part)
    
    def __str__(self):
        return (f"{self.real}+{self.imaginary}i")
    
if __name__=="__main__":
    c1=ComplexNumber(2,3)
    c2=ComplexNumber(1,4)

    result_add=c1+c2
    print(f"Addition:{result_add}")

    result_sub=c1-c2
    print(f"Substraction:{result_sub}")

    result_mul=c1*c2
    print(f"multiplication:{result_mul}")
```

```python
#7
import numpy as np
import time

# Number of elements
size = 1000

# Python list operation
python_list = list(range(size))
start_time = time.time()
python_list_result = [x + 1 for x in python_list]
end_time = time.time()
python_list_time = end_time - start_time

# NumPy array operation
numpy_array = np.arange(size)
start_time = time.time()
numpy_array_result = numpy_array + 1
end_time = time.time()
numpy_array_time = end_time - start_time

# Print results
print(f"Time taken using Python list: {python_list_time:.6f} seconds")
print(f"Time taken using NumPy array: {numpy_array_time:.6f} seconds")
if numpy_array_time > 0:  # To avoid division by zero
    print(f"NumPy is approximately {python_list_time / numpy_array_time:.2f} times faster for the operation")
else:
    print("NumPy was so fast that the time measured was effectively zero.")
```

```python
#8
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

# Sample data
data = {
    'Variable_A': [1, 2, 3, 4, 5, 6],
    'Variable_B': [1.9, 3.4, 5.6, 10, 3.3, 7.5]
}

df = pd.DataFrame(data)

# 1. Calculate correlation
def calculate_correlation(df, col1, col2):
    correlation, _ = pearsonr(df[col1], df[col2])
    return correlation

# 2. Calculate joint probability
def joint_probability(df, event_A, variable_A, event_B, variable_B):
    joint_df = df[(df[variable_A] == event_A) & (df[variable_B] == event_B)]
    joint_prob = len(joint_df) / len(df)
    return joint_prob

# 3. Calculate marginal probability
def marginal_probability(df, event, variable):
    marginal_df = df[df[variable] == event]
    marginal_prob = len(marginal_df) / len(df)
    return marginal_prob

# Example usage
correlation = calculate_correlation(df, 'Variable_A', 'Variable_B')
print(f"Correlation: {correlation}")

# Example joint and marginal probability (replace events with appropriate values)
event_A = 3
event_B = 5.6
joint_prob = joint_probability(df, event_A, 'Variable_A', event_B, 'Variable_B')
print(f"Joint Probability (A=3, B=5.6): {joint_prob}")

marginal_prob_A = marginal_probability(df, event_A, 'Variable_A')
print(f"Marginal Probability (A=3): {marginal_prob_A}")

marginal_prob_B = marginal_probability(df, event_B, 'Variable_B')
print(f"Marginal Probability (B=5.6): {marginal_prob_B}")
```

```python
#9
import pandas as pd

# Sample data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', None, 'Que'],
    'Age': [25, None, 30, 99, None],
    'City': ['New York', 'Los Angeles', None, 'Chicago', 'Miami']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Function to analyze missing data
def analyze_missing_data(df):
    print("Dataframe:")
    print(df)
    
    print("\nMissing Data Analytics:")
    
    # Check for missing values
    missing_data = df.isnull().sum()
    print("\nMissing values in each column:")
    print(missing_data)
    
    # Calculate percentage of missing values
    missing_percentage = (missing_data / len(df)) * 100
    print("\nPercentage of missing values in each column:")
    print(missing_percentage)
    
    return missing_data, missing_percentage

# Analyze missing data
missing_data, missing_percentage = analyze_missing_data(df)

# Example of handling missing data by filling placeholders
df_filled = df.fillna("Unknown")
print("\nDataFrame after handling missing data:")
print(df_filled)

# Function to handle missing data
def handle_missing_data(df, choice='1'):
    print("\nChosen method to handle missing data:", choice)
    
    if choice == '1':
        value = "Unknown"  # Example placeholder value
        df_filled = df.fillna(value)
        print("\nData after filling missing values:")
        print(df_filled)
        return df_filled
    
    elif choice == '2':
        df_dropped_rows = df.dropna()
        print("\nData after dropping rows with missing values:")
        print(df_dropped_rows)
        return df_dropped_rows

    elif choice == '3':
        df_dropped_columns = df.dropna(axis=1)
        print("\nData after dropping columns with missing values:")
        print(df_dropped_columns)
        return df_dropped_columns
    
    else:
        print("Invalid choice! Using default option (1).")
        return df.fillna("Unknown")

# Example usage: handle missing data with a predefined choice
df_handled = handle_missing_data(df, choice='1')  # You can change '1' to '2' or '3' to test other options
```

```python
#10
import pandas as pd
import matplotlib.pyplot as plt
data = {
'age':[25,30,35,40,45,50,55,60,65,70],
'salary':[20000,25000,30000,35000,40000,45000,50000,55000,60000,65000],
'department':['hr','it','finance','it','finance','hr','hr','finance','it','finance']
}
df=pd.DataFrame(data)

plt.figure(figsize=(8,5))
plt.scatter(df['age'],df['salary'],color='blue',marker='o')
plt.title("Scatter plot of age vs salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.show()


plt.figure(figsize=(8,5))
plt.hist(df['age'],bins=5,color='green',edgecolor='black')
plt.title("Histogram of age distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()


plt.figure(figsize=(8,5))
department_counts=df['department'].value_counts()
plt.bar(department_counts.index,department_counts.values,color='orange',edgecolor='black')
plt.title("Bar plot of department_count")
plt.xlabel("Department")
plt.ylabel("Number of` employees")
plt.grid(axis='y')
plt.show()
```

```python
#11
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
'age':[25,30,35,40,45,50,55,60,65,70],
'salary':[20000,25000,30000,35000,40000,45000,50000,55000,60000,65000],
'department':['hr','it','finance','it','finance','hr','hr','finance','it','finance']
}
df=pd.DataFrame(data)
sns.set(style="whitegrid")


plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='age',y='salary',hue='department',palette='viridis',s=100)
plt.title("Scatter plot of age vs salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(title='Department')
plt.show()


plt.figure(figsize=(8,5))
sns.histplot(data=df,x='age',bins=5,kde=True,color='purple')
plt.title("Histogram of age distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()


plt.figure(figsize=(8,5))
sns.countplot(data=df,x='department',palette='pastel')
plt.title("Bar plot of department_count")
plt.xlabel("Department")
plt.ylabel("Number of employees")
plt.show()
```
